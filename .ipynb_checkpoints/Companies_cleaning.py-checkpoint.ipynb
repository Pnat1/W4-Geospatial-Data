```python
by_cities = []
for i in range(df_ESP.shape[0]):
    try:
        if df_ESP["offices"][i][0]["city"] in ["Madrid"]:
            by_cities.append(df_ESP["name"][i])
            df_ESP_mad = df_ESP.loc[df_ESP['name'].isin(by_cities)]

    except:
        pass   
    
by_cities = []
for i in range(df_GBR.shape[0]):
    try:
        if df_GBR["offices"][i][0]["city"] in ["London"]:
            by_cities.append(df_GBR["name"][i])
            df_GBR_ldn = df_GBR.loc[df_GBR['name'].isin(by_cities)]

    except:
        pass   

by_cities = []
for i in range(df_DEU.shape[0]):
    try:
        if df_DEU["offices"][i][0]["city"] in ["Berlin"]:
            by_cities.append(df_DEU["name"][i])
            df_DEU_ber = df_DEU.loc[df_DEU['name'].isin(by_cities)]

    except:
        pass   
```

```python
df_ESP_mad = df_ESP_mad.reset_index().drop("index",1)
df_GBR_ldn = df_GBR_ldn.reset_index().drop("index",1)
df_DEU_ber = df_DEU_ber.reset_index().drop("index",1)
```

```python
# We filter by number of employees now with a range between 10 and 200 
# Considering the companies dataframe is made of tech companies we do not care about other conditions
# Porque no se pueden hacer con variables

t = []
for j in [df_DEU_ber, df_ESP_mad, df_GBR_ldn]:
    
    by_employees = []

    for i in range(j.shape[0]):
        if 10 < j["number_of_employees"][i] < 160:
            by_employees.append(j["name"][i])
    
    t.append(by_employees)

    
```
